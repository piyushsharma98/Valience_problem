# Visualisation from Rural Health Statistics 2016



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Statistics of State/Ut-wise Physicians at CHCs as on 31st March, 2016

```python
physicians = pd.read_csv("Table_26_State_Uts.csv")
physicians
```

# Required Physicians as on 31st March, 2016

```python
physicians.plot(kind="barh", x="State/UT", y="Required - [R] - (As on 31st March, 2016)")
plt.xlabel('Required  Physicians')
plt.show()
```

# Sanctioned Physicians as on 31st March, 2016

```python
physicians.plot(kind="barh", x="State/UT", y="Sanctioned - [S] - (As on 31st March, 2016)")
plt.xlabel('Sanctioned Physicians')
plt.show()
```

# In Position Physicians as on 31st March, 2016

```python
physicians.plot(kind="barh", x="State/UT", y="In Position - [P] - (As on 31st March, 2016)")
plt.xlabel('In Position Physicians')
plt.show()
```

# Vacant Physicians as on 31st March, 2016

```python
physicians.plot(kind="barh", x="State/UT", y="Vacant - [S-P] - (As on 31st March, 2016)")
plt.xlabel('Vacant Physicians')
plt.show()
```

# Shortfall Physicians as on 31st March, 2016

```python
physicians.plot(kind="barh", x="State/UT", y="Shortfall - [R-P] - (As on 31st March, 2016)")
plt.xlabel('Shortfall Physicians')
plt.show()
```

# Total of Sanctioned, In Position, Vacant Physicians as on 31st March, 2016

```python

total = {'vacancy' : ['Sanctioned','In Position','Vacant'],
         'Numbers' :[physicians['Sanctioned - [S] - (As on 31st March, 2016)'].sum(),
                     physicians['In Position - [P] - (As on 31st March, 2016)'].sum(),
                     physicians['Vacant - [S-P] - (As on 31st March, 2016)'].sum()
                    ]
         }
```

```python
data_frame = pd.DataFrame(total)
data_frame
```

```python
data_frame.plot(kind="bar", x="vacancy", y="Numbers")
plt.xlabel('vacancy')
plt.show()
```

```python
labels = 'Sanctioned', 'In Position', 'Vacant'
nums = data_frame['Numbers']
colors = ['lightgreen', 'lightblue', 'lightcoral']
explode = (0, 0.3, 0.1)

plt.pie(nums, explode=explode, labels=labels, colors=colors,autopct='%1.2f%%', shadow=True)
plt.axis('equal')
plt.show()
```
